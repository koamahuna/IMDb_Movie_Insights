Now that we have prepped our data for analysis we can begin to answer our main question: how can we improve a film's profitabilty?
```python
# We decided to create an intermediary copy of our orginal dataframe so as not to overwrite any important
#columns during our analysis.
jdf = df
```

```python
#Cleaning the data to make the analysis easier, as there are a significant amount of null values.
#We found there is not much research online about a standard for a dropping percentage threshold some say 25-30 others say 85-95, others 70.
col_na_percent = jdf.isnull().sum()/len(jdf) * 100
col_na_percent.sort_values(ascending=False)
#Keeping all remaining columns because we plan on using movie facebook likes in the analysis, despite it's null ratio.

row_na_percent = jdf.isnull().sum(axis=1)/len(jdf.columns)*100
row_na_percent.sort_values(ascending=False) #We decided to use 30% percent based on previous class work, using that as our standard.
jdf = jdf.loc[row_na_percent < 30]

#As you can see it shrunk from 3807 rows to 3789 rows so it did not shrink our data by too much.

#For the remainder of null numerical values we can use median because it won't significantly affect data
#Mean would not be as reliable because we have outliers.
jdf.fillna((jdf.median()), inplace = True)

#We could fill non-numerical values with the value from the previous row.
#We chose not to use this code bc it wont work with the data we are using. 
#It will give movies incorrect info and could skew data and our analysis
#jdf.fillna(method='ffill', inplace = True)

#In order to eliminate more outliers we set sliced the data to only include movies with a title year after 1950.
#These movies came out at a time where the box office was not as big, so there values are misleading
jdf = jdf.loc[jdf.title_year > 1950]

jdf.info() #most nulls removed or replaced
```

```python
jdf['genres'] = jdf['genre_1'] +', ' + jdf['genre_2'] +', ' + jdf['genre_3']
#Creating two columns (profit and ROI) for a stronger anaylysis than just gross.
jdf['profit'] = jdf.gross - jdf.budget

jdf['ROI'] = (jdf.profit/jdf.budget)*100
```

```python
#Finding the most successful genres at the domestic box office
top_box_office = jdf.groupby('genres').gross.sum().nlargest(10)

#Creating a visual to better understand
top_box_office.sort_values(ascending=True, inplace=True)
clrs = ['grey' if (x < max(top_box_office)) else 'green' for x in top_box_office]
top_box_office_bar = plt.barh(top_box_office.index, width=top_box_office,color = clrs)
plt.xlabel('USD (in millions)')
plt.yticks(label=top_box_office.index, fontsize = 12)
plt.title('Sum Gross at Domestic Box Office')
plt.show()

#plt.savefig('Installs_byCat_bar.png', dpi=300, bbox_inches='tight')
```

Genre analysis:

Because genre is non numerical, we cannot determine its correlation with gross, profit, or ROI, but we are still assuming it plays a major role in determining the success of a movie, so we wanted to make it a core part of our analysis. Action|Adventure|Sci-Fi movies are the highest grossing genre for the domestic box office. This would include movies like The Avengers, so it's no shock why they're are at the top, as these movies are typically released as tentpoles or summer blockbusters, which appeal to the widest audience and draw in the largest crowds. Following after are various versions of comedies, dramas and action movies, which are also typically associated with blockbusters.

```python
#Finding the least successful genres at the domestic box office
lowest_box_office = jdf.groupby('genres').gross.sum().nsmallest(10)

#Creating a visualization
lowest_box_office.sort_values(ascending=False, inplace=True)
lowest_box_office_bar = plt.barh(lowest_box_office.index, width=lowest_box_office,color = "blue")
plt.xlabel('USD (in millions)')
plt.yticks(label=lowest_box_office.index, fontsize = 12)
plt.title('Least Successful Genres at the Domestic Box Office')
plt.show()
```

Several factors could have contributed to these movies' failures, but it's safe to say these are not winning combinations.

```python
#Finding the most expensive genres
top_budget = jdf.groupby('genres').budget.mean().nlargest(10)

#Creating a visual
top_budget.sort_values(ascending=True, inplace=True)
clrs = ['grey' if (x < max(top_budget)) else 'red' for x in top_budget]
top_budget_bar = plt.barh(top_budget.index, width=top_budget,color = clrs)
plt.xlabel('USD (in 100 millions)')
plt.yticks(label=top_budget.index, fontsize = 12)
plt.title('Most Expensive Genres')
plt.show()
```

On average, we can see combinations of animation, action, adventure, and sci-fi movies tend to be the most expensive. This is likely do due to big and extravegant sets or shooting locations, big-time actors, and VFX. It's also worth noting these movies are closely similar to those that make up the largest segment of the box office.

```python
#Finding the least expensive genres
smallest_budget = jdf.groupby('genres').budget.mean().nsmallest(10)

#Creating a visual
smallest_budget.sort_values(ascending=False, inplace=True)
clrs = ['green' if (x == min(smallest_budget)) else 'grey' for x in smallest_budget]
smallest_budget_bar = plt.barh(smallest_budget.index, width=smallest_budget,color = clrs)
plt.xlabel('USD (in millions)')
plt.yticks(label=smallest_budget.index, fontsize = 12)
plt.title('Least Expensive Genres')
plt.show()
```

```python
#Comparing the most and the least profitable genres with a visual
least_profit = jdf.groupby('genres').profit.mean().nsmallest(8)
least_profit.sort_values(ascending=True, inplace=True)
top_profit = jdf.groupby('genres').profit.mean().nlargest(8)
top_profit.sort_values(ascending=True, inplace=True)

least_profit_bar = plt.barh(least_profit.index, width=least_profit,color = "red")
top_profit_bar = plt.barh(top_profit.index, width=top_profit,color = "green")
plt.xlabel('USD (in millions)')
plt.yticks(label=least_profit.index, fontsize = 12)
plt.title('Most & Least Profitable Genres')
plt.show()
```

Surpsingly, what we see is that similar genres make up the most and least profitable genres. This could be an indication that genre is not as closely linked to success as we thought.

```python
#Finding the smartest investments by genre
top_ROI = jdf.groupby('genres').ROI.mean().nlargest(10)

#Creating a visualization
top_ROI.sort_values(ascending=True, inplace=True)
clrs = ['grey' if (x < max(top_ROI)) else 'green' for x in top_ROI]
top_ROI_bar = plt.barh(top_ROI.index, width=top_ROI,color = clrs)
plt.xlabel('USD (in millions)')
plt.yticks(label=top_ROI.index, fontsize = 12)
plt.title('Highest ROI by Genre')
plt.show()
```

```python
#Critics' reception toward genre
critics = jdf.groupby('genres').imdb_score.mean().nlargest(10)
critics.sort_values(ascending=True, inplace=True)

critics_bar = plt.barh(critics.index, width=critics,color = "blue")
plt.xlabel('Imdb Score)')
plt.yticks(label=critics.index, fontsize = 12)
plt.title("Critics' Favorite Genre")
plt.show()
```

```python
 # Theoretically, 25 to 30% is the maximum missing values are allowed, 
    # beyond which we might want to drop the variable from analysis.
#get rid of the columns that have more than 30% values as null
df.isnull().sum()/len(df) * 100
#column have 30%+ null val: 
    #genre_3(keep),genre_4,genre_5,genre_6,genre_7,genre_8,
    #facenumber_in_poster,movie_facebook_likes(keep),genres(keep) 

df2=df.drop(columns=['genre_4',
                     'genre_5',
                     'genre_6',
                     'genre_7',
                     'genre_8'],axis=1)
df2.info()
```

We decided to add a profit column in order to more accurately estimate the perforamnce of a film. We created this column by subtracting budget from revenue. Finally, we check to see if any movies are recording negative profit. After adding the profit column, we used Python to check for the percentage of missing values for each row and displayed the result to give a brief look. Next, I dropped the rows that have more than 35% of missing values in df2. Before I drop them, there were 5043 rows in total, and there are 5004 rows now. Finally, in business people generally use both ROI and profit margin to measure their profitability. Neither one is "better" in some absolute sense but depends on the questions and conditions. For instance, ROI can help you understand the result of the added investment but profit margin can help you figure out how much profit your company brings. As a result, I created two columns for net profit margin/profit margin and ROI, using the profit, gross(revenue), and budget columns from df2.

```python
#add profit column to df2
df2['profit'] = df2['gross'] - df2['budget']

#make sure that there's no negative numbers for profit 
df2.loc[df2['profit'] < 0 , 'profit'] = 0
```

```python
nullratio_row=df2.isnull().sum(axis=1)/len(df2.columns) * 100
nullratio_row.sort_values(ascending=False)
#drop rows with more than 35% missing values
df2 = df2.loc[nullratio_row <= 35]
#before:5043 rows; now:5004 rows
```

```python
#get net profit margin:Net Profit Percentage = Net Profit/Total Sales * 100
df2['NPM'] = (df2.profit/df2.gross) * 100
#get ROI (Return On Investment): ROI = (profit/budget)*100
df2['ROI'] = (df2.profit/df2.budget) * 100
```

```python
#building correlation matrix for df2
df2corr=df2.corr()
df2corr
#visualize correlation matrix 
plt.figure(figsize=(14,14))
sns.heatmap(df2corr,annot=True,linewidths=.5)
plt.title("Correlation Matrix of IMDB Data")
plt.show()
```

Consequently, to interpret this graph, each cell in the grid represents the value of the correlation coefficient betweeen two variables.

Each cell in the grid represents the value of the correlation coefficient between two variables.

All diagonal elements are 1, since they are represent the correlation of each variable with itself.

A large positive value (near to 1.0) indicates a strong positive correlation.

A large negative value (near to -1.0) indicates a strong negative correlation (Analyse-It, 2020).

There are 3 degrees when describing the density of the relationship: high, moderate, and low.

From the heatmap of correlation matrix, we can find that there are 4 high degree linear relationships and 4 moderate degree linear relationships.

High: profit-> gross, num_voted_users,NPM

NPM -> profit

Moderate: profit ->num_critic_for_reviews, num_user_for_reviews,movie_facebook_likes

NPM -> gross

The rest have a low degree of correlation, so I decided to not put them into account to our regression model.

High correlation is defined as a coefficient value that lies between ± 0.50 and ± 1 for our purposes. Moderate degree: If the value lies between ± 0.30 and ± 0.49, then it is said to be a medium correlation. Low degree: When the value lies below + .29, then it is said to be a small correlation (StatisticsSolutions, 2020).

Since profit and NPM are contributed by gross and budget, they won't be used as dependent variables in the model.

The relationship between ROI and all other variables are all very weak so ROI won't be considered as the independent variable.

The only two relationships have with NPM (Net Profit Margin) are all correlated with profit and gross which is unnecessary to put into predicted model.

In conclusion, profit will be used as the dependent variable for the regression model.

To get the regression model, we first need to fill all the null values. After consideration, I decided to fill all the null values with the median of its columns.

Therefore, I replaced missing values with the median for the columns of num_voted_users, num_critic_for_reviews, num_user_for_reviews,movie_facebook_likes, and profit. And then, double check to make sure there's no remaining missing values for the selected independent variables.

```python
import math
#fillna for num_voted_users
median_voted = math.floor(df2.num_voted_users.median())
median_voted #34791
df2.num_voted_users = df2.num_voted_users.fillna(median_voted)

#fillna for num_critic_for_reviews
median_critic = math.floor(df2.num_critic_for_reviews.median())
median_critic #111
df2.num_critic_for_reviews = df2.num_critic_for_reviews.fillna(median_critic)

#fillna for num_user_for_reviews
median_user = math.floor(df2.num_user_for_reviews.median())
median_user #157
df2.num_user_for_reviews = df2.num_user_for_reviews.fillna(median_user)

#fillna for movie_facebook_likes
median_facebook = math.floor(df2.movie_facebook_likes.median())
median_facebook #1000
df2.movie_facebook_likes = df2.movie_facebook_likes.fillna(median_facebook)

#fillna for profit 
median_profit = math.floor(df2.profit.median())
median_profit #902679
df2.profit = df2.profit.fillna(median_profit)
```

Finally, I used linear_model function from the sklearn package to build a regression model for df2, with profit column as our independent variable and 4 columns as our dependent variables.

```python
#import pandas as pd 
#import numpy as np 
from sklearn import linear_model

reg = linear_model.LinearRegression()
#dependent variable: profit; 
#independent variables: 
    #num_voted_users,num_critic_for_reviews,num_user_for_reviews,movie_facebook_likes

reg.fit(df2[['num_voted_users',
             'num_critic_for_reviews',
             'num_user_for_reviews',
             'movie_facebook_likes']],df2.profit)
```

Our predictive model offers some interesting insights. Firstly, if you add one more user vote, the profit of the movie will be increased by 167 dollars.
